```
from ipa import Ipa

import time

drive = '/content/drive/My Drive/ml/ipa/data'

train_dir = drive+'/train'
test_dir = drive+'/test'
score_dir = drive+'/score'

ipa = Ipa()

```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
start_time = time.time()
ipa.train(train_dir)
print("--- train %s seconds ---" % (time.time() - start_time))
```

```
start_time = time.time()
results = ipa.predict(test_dir)
results.to_csv('test_scores',index=False)
print("--- test %s seconds ---" % (time.time() - start_time))
```

```
start_time = time.time()
results = ipa.predict(score_dir)
results = results.drop(['Labels'], axis=1)
results.to_csv('scores',index=False)
print("--- predict %s seconds ---" % (time.time() - start_time))
```
